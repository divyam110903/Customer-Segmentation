```python
import pandas as pd

# Load the dataset
file_path = 'Mall_Customers.csv'
data = pd.read_csv('/content/Mall_Customers.csv')

# Display the first few rows of the dataset
data.head(10)
```

Data transformation


```python
data.columns = ["CustomerID", "Gender", "Age", "AnnualIncome", "SpendingScore"]
print(data)

data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})
```

```python
data.describe()
```


# **Visualizations of age distribution, annual income distribution, and spending score distribution.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=30, kde=True,color='green')
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['AnnualIncome'], bins=30, kde=True)
plt.title('Annual Income Distribution')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['SpendingScore'], bins=30, kde=True, color='purple')
plt.title('Spending Score Distribution')
plt.show()
```

# Visualizing relationships
A scatter plot showing the relationship between annual income and spending score


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='AnnualIncome', y='SpendingScore', hue='Gender')
plt.title('Income vs Spending Score')
plt.show()
```
